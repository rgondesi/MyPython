```python
import glob, os

#os.chdir("DataFiles")
with open("happynumbers.txt") as happyNumFile:
    linetext = happyNumFile.read()
    happyNumFile_List = linetext.split()
    #print(happyNumFile_List)

with open("primenumbers.txt") as primenumbersFile:
    linetext = primenumbersFile.read()
    primenumbersFile_List = linetext.split()
    #print(primenumbersFile_List)
    
NewList = []
    
for item1 in happyNumFile_List:
    if item1 not in NewList:
        NewList.append(item1)
        
for item2 in primenumbersFile_List:
    if item2 not in NewList:
        NewList.append(item2)

print(NewList)
```
